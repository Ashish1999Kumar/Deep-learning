```
import numpy as np
import pandas as pd  
```

```
import matplotlib.pyplot as plt
```

```
import sklearn.datasets as d
```

```
breast_cancer=d.load_breast_cancer()#loading a predefined datasets breast_cancer
```

```
X=breast_cancer.data
Y=breast_cancer.target
```

```
data=pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)#storing the value of the datasets in 
```

```
data['class']=Y
```

```
from sklearn.model_selection import train_test_split as tt
```

```
X=data.drop('class',axis=1)
```

```
Y=data.filter(['class'],axis=1)
```

```
X_train,X_test,Y_train,Y_test=tt(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```
X_test.shape
```

```
plt.plot(X_test.T,"*")
plt.xticks(rotation='vertical')
```

```
X_bins=X_train.apply(pd.cut,bins=2,labels=[1,0])
```

```
X_bi=X_test.apply(pd.cut,bins=2,labels=[1,0])
```

```
plt.plot(X_bins.T,"*")
plt.xticks(rotation='vertical')
```

```
X_bins=X_bins.values
X_bi=X_bi.values
```

```
Y_bins=Y_train.values
Y_bi=Y_test.values
```

```
for b in range(0,X_bins.shape[1]):
  Y_pred=[]
  acc=0
  for x,y in zip(X_bins,Y_bins):
    Y_pred.append(np.sum(x)>=b)
    acc+=((np.sum(x)>=b)==y)
  print(acc,acc/X_bins.shape[0],b)
```

```
X_train.shape
```

```
from sklearn.metrics import accuracy_score
```

```
b=27
i=0
Y_pred_test=[]
for x in X_bi:
  Y_pred=np.sum(x)>=b
  Y_pred_test.append(Y_pred)
acc=accuracy_score(Y_pred_test,Y_bi)
for y in Y_pred_test:
  print(y,i)
  i=i+1
print(b,acc)
```
