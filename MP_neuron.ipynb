```
from random import randint
```

```
i=randint(0,X_bin_train.shape[0])
b=3
print('for row ',i)
if(np.sum(X_bin_train[i,:])>=b):
  print("like")
else:
  print("not_like")
if(Y_train[i]==1):
  print("like")
else:
  print("not_like")
```

```
for b in range(0,X_bin_train.shape[1]):
  Y_pred_train=[]
  acc=0
  for x,y in zip(X_bin_train,Y_train):
    Y_pred=np.sum(x)>=b
    Y_pred_train.append(Y_pred)
    acc+=(Y_pred==y)
  print(acc,acc/X_bin_train.shape[0],b)
```

```
from sklearn.metrics import accuracy_score
```

```
b=28
i=0
Y_pred_test=[]
for x in X_bin_train:
  Y_pred=np.sum(x)>=b
  Y_pred_test.append(Y_pred)
acc=accuracy_score(Y_pred_test,Y_test)
for y in Y_pred_test:
  print(y,i)
  i=i+1
print(b,acc)
```
